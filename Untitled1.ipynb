```python
#average bagging all of the submissions
import pandas as pd
import numpy as np
```

```python
files = []
filenames = ['submit_1.csv', 'submit_2.csv', 'submit_3.csv', 'submit_6.csv']
```

```python
for filename in filenames:
    files.append(pd.read_csv(filename))
```

```python
files[0]
```

```python
final_sub = pd.DataFrame()
final_sub['PERID'] = files[0]['PERID']
final_sub['Criminal'] = files[0]['Criminal']
```

```python
for i in range(len(files)):
    final_sub['Criminal'] += files[i]['Criminal']
```

```python
new_final_sub = final_sub.copy()
```

```python
new_final_sub['Criminal'] = new_final_sub['Criminal'].apply(lambda x : 1 if x >= 1 else 0)
```

```python
new_final_sub.to_csv("final_sub_2.csv", index = False)
```
